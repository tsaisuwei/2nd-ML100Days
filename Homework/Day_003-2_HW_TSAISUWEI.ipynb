```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[1:100]
print(data) 
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
import pandas as pd
split_tag = '\t'


data_arrange= data.split(split_tag) #全部用\N切開  用","分開
print(len(data))  #列出有幾行
print(data)   #列出全部
data[100] # 



#data = data.split(split_tag_1,split_tag_2) #全部用\t切開  用","分開


data_arrange
```
